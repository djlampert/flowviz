## Vector Plotting in Python
This is an example of how to plot fluid vector fields in Python. The objective is to plot the vector velocity and material derivative for the function:  

$\vec{V}=(0.5 + 0.8x)\hat{i}+(1.5-0.8y)\hat{j}$  

The image below is the solution, which is Figure 4-14 from *Fluid Mechanics: Fundamentals and Applications, 1st Edition*, by Cembel and Cingala:
![title](https://i.ibb.co/q139MhJ/image.png)

```python
# use numpy to make a grid of values
import numpy
xs = numpy.linspace(-2,2,5)
ys = numpy.linspace(0,5,6)
print(xs,ys)
```

```python
# make the grid
gxs, gys = numpy.meshgrid(xs,ys)
print(gxs)
print(gys)
```

```python
# calculate the velocity in x and y directions
u = 0.5+0.8*gxs
v = 1.5-0.8*gys
print(u)
```

```python
# use matplotlib to make a plot
import matplotlib.pyplot as plt

# use quiver to make the velocity vectors
p1 = plt.quiver(gxs,gys,u,v, scale = 20, color = 'deeppink')

# use streamplot to add flowlines for the field
p2 = plt.streamplot(gxs,gys,u,v, color = 'k', density = 0.4, linewidth = 0.5)

# add a grid, set limits, add labels, make aspect equal
plt.grid(lw = 0.5)
plt.xlim(-3,3)
plt.ylim(-1,5.6)
plt.gca().set_aspect('equal')
plt.xlabel('x', weight = 'bold')
plt.ylabel('y', weight = 'bold')
plt.title('Velocity Field', weight = 'bold')

# add the stagnation point
p3 = plt.plot(-0.625,1.875, 'o', markersize = 10, markeredgecolor = 'deeppink',
              markerfacecolor = 'None')

# optionally save the image to a file by uncommenting the line below
# plt.savefig('example_vector_plot', dpi = 300)
```

```python
# calculate the material acceleration fields
ax=0.4+0.64*gxs
ay=-1.2+0.64*gys
```

```python
# use quiver to make the vectors
p1 = plt.quiver(gxs,gys,ax,ay, scale = 20, color = 'deeppink')

# use streamflow to add flow lines to field
p2 = plt.streamplot(gxs,gys,u,v, color = 'k', density = 0.4, linewidth = 0.5)

# add a grid, set limits, add labels, make aspect equal
plt.grid(lw = 0.5)
plt.xlim(-3,3)
plt.ylim(-1,5.6)
plt.gca().set_aspect('equal')
plt.xlabel('x', weight = 'bold')
plt.ylabel('y', weight = 'bold')
plt.title('Material Acceleration Field and Streamlines', weight = 'bold')

# add the stagnation point
p3 = plt.plot(-0.625,1.875, 'o', markersize = 8, markeredgecolor = 'deeppink',
              markerfacecolor = 'None')
```
